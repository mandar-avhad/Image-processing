```
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```
from google.colab.patches import cv2_imshow
```

```
im = cv2.imread('/content/bag4_view1.jpeg')
imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
```

```
ret,thresh = cv2.threshold(imgray,90,255,0)
plt.imshow(thresh)
cv2.imwrite("new.jpg",thresh)
```

```

dil = cv2.dilate(thresh, None, iterations=8)
```

```
plt.imshow(dil)
```

```
contours, hierarchy = cv2.findContours(dil,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```
cv2.drawContours(im, contours, -1, (0, 255, 0), 3)
plt.imshow(im)
```

```
cnt = sorted(contours, key=cv2.contourArea, reverse=True)
```

```
cv2.drawContours(im, cnt, 1, (0, 255, 0), 3)
plt.imshow(im)
```

```

rect = cv2.boundingRect(cnt[1])
print(cv2.contourArea(cnt[1]))
x,y,w,h = cv2.boundingRect(cnt[1])
```

```
cv2.rectangle(im,(x,y),(x+w,y+h),(0,255,0),2)
cv2.putText(im,'breadth: '+str(w),(x-10, y+h+10),0,0.3,(0,255,0))
cv2.putText(im,'height: '+str(h),(x+w-50, y-10),0,0.3,(0,255,0))
```

```
cv2_imshow(im)
```

```
cv2.imwrite("res_b4_view1.jpeg",im)
```

```
print(x,y,w,h)
```
